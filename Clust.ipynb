```python
with open('./ahd-dictionary/ohg_dict_final.html') as f:
    DICT = json.loads('['+','.join(f.readlines())+']')
```

```python
from collections import defaultdict
```

```python
CLUST = {}
for i in tqdm(range(0, data.shape[0])):
    for j in tqdm(data.iloc[i].dropna().values[1:], leave=False):
        if j is not None:
            CLUST[j] = data.iloc[i][0]
```

```python
CLUST['demo']
```

```python
json.dump(CLUST, open('lemmas.json', 'w'), ensure_ascii=False)
```

```python
CLUST = defaultdict(list)
for i in tqdm(range(0, data.shape[0])):
    for j in tqdm(data.iloc[i].dropna().values[1:], leave=False):
        for k in tqdm(DICT, leave=False):
            if data.iloc[i][0] in k['headers']:
                CLUST[j].append(k['entry'])
```

```python
json.dump(CLUST, open('tooltip_dict.json', 'w'), ensure_ascii=False)
```
